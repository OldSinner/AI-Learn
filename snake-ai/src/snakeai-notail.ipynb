Zaimportowanie niezbędnych bibliotek i narzędzi do pracy z danymi oraz modelami uczenia maszynowego.


```python
import pygame
import random
from collections import defaultdict
import numpy as np
from numpy import argmax
import matplotlib.pyplot as plt
import statistics
from snakenotail import SnakeAI
from basesnake import get_direction_from_keyboard
snake = SnakeAI()

try:
    while snake.running:
        action = get_direction_from_keyboard()
        _,_,done = snake.step(action)
        if done:
            print("ugabuga")
            snake.game_reset()
        if not snake.running:
            pygame.quit()
finally:
    pygame.quit()
```
